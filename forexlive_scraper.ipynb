# Extract Forexlive URL from Scraped Twitter Data (splits data into monthly)

```python
"""
Input:
  'forexlive-2022.csv'
  'forexlive-2021.csv'
  'forexlive-2020.csv'
  'forexlive-2019.csv'
  etc

Output:
  'clean_forexlive-2010_1
  'clean_forexlive-2010_2'
  'clean_forexlive-2010_3'
  etc
"""

csv = 'forexlive-2022.csv'
# csv = 'forexlive-2021.csv'
# csv = 'forexlive-2020.csv'
# csv = 'forexlive-2019.csv'
# csv = 'forexlive-2018.csv'
# csv = 'forexlive-2017.csv' # error parsing this one, go into the file and manually fix
# csv = 'forexlive-2016.csv'
# csv = 'forexlive-2015.csv'
# csv = 'forexlive-2014.csv'
# csv = 'forexlive-2013.csv'
# csv = 'forexlive-2012.csv'
# csv = 'forexlive-2011.csv'
# csv = 'forexlive-2010.csv'
# csv = 'forexlive-2009.csv'

df = pd.read_csv(csv, error_bad_lines=False) # quoting=3, 


pd.set_option('max_colwidth', 800)
df.loc[df["urls"] == "[]", "urls"] = 'http' + df['tweet'].str.split('http', expand=True)[1] # replace urls column with tweet column if blank
df.loc[df['urls'].str[0:1] == "[", "urls"] = df['urls'].str[2:-2] # if bracketted, remove brackets
df['urls'] = df['urls'].str.replace(' #MKT', '') # if there is a hashtag #MKT in the text, remove
df['tweet'] = df['tweet'].str.replace('ForexLive: ', '') # remove tag for 2011
df['tweet'] = df['tweet'].str.replace('New blog post: ', '') # remove tag for 2010 - 2009
df['body']=''
df[['created_at', 'date', 'time', 'timezone', 'tweet','urls','body']]

# split into monthly csv for scraping monthly
for i in reversed(range(1,13)):
  df2 = df.loc[df['date'].str[5:7].astype('int32') == i]
  df2[['created_at', 'date', 'time', 'timezone', 'tweet','urls','body']].to_csv('clean_'+csv[:-4]+'_'+str(i)+'.csv', sep=',', index=False)
  files.download('clean_'+csv[:-4]+'_'+str(i)+'.csv')

# # # split into yearly csv for scraping yearly
# df[['created_at', 'date', 'time', 'timezone', 'tweet','urls','body']].to_csv('clean_'+csv[:-4]+'.csv', sep=',', index=False)
```

# Scrape Forexlive From URLs (monhtly)

```python
"""
This cell is meerely a function that ocntains the web scraper
it is called from the cell below
"""


!pip install kora -q
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from bs4.element import Comment
import urllib.request
import time
from tqdm import tqdm
import requests
from kora.selenium import wd
from google.colab import files, output

def scrape_forexlive(csv):
  """ Scrapes Forexlive.com pages for articleBody, dateline, headline, articleSection, datePublished, genre and name
  Args:
    csv: name of local csv file containing url data to scrape
  Returns:
    No Return output, but saves scraped data csv to disk and beeps when each month is complete
  """
  df = pd.read_csv(csv) # , skip_blank_lines=True
  df['url_2'] = ''
  df.rename(columns = {'body':'articleBody'}, inplace = True)
  df['articleBody'] = df['articleBody'].astype("string")
  df['dateline'] = ''
  df['headline'] = ''
  df['articleSection'] = ''
  df['datePublished'] = ''
  df['genre'] = ''
  df['name'] = ''

  # when DMing, there is no URL to scrape so might as well delete the row
  # df['urls'].replace(np.nan, 'No_URL', inplace=True)
  df.dropna(subset = ['urls'], inplace = True)
  df = df.reset_index()

## 2 - def
  for i in tqdm(range(len(df))):

#### 4 
    termcount = 0
    url = df.loc[i]['urls']

    ############################################################################################################
    # this section if for the  working URLS
    ############################################################################################################

#### 4 -  for i in tqdm(range(len(df))):
    if df.loc[i]['urls'].find('.forexlive.com/')>0 or df.loc[i]['urls'].find('bit.ly')>0 or df.loc[i]['urls'].find('/disq.us/')>0 or df.loc[i]['urls'].find('/go.shr.lc/')>0 or df.loc[i]['urls'].find('/ow.ly/')>0: 
      
###### 6   
      try:   
        response = requests.get(df.loc[i]['urls'], timeout=240)
        if response.status_code==200: # if return successful
  ######## 8
          try:
            wd.get(url)
            html = wd.page_source
            html = html.split('<script charset="utf-8" src="')[0]
            html = html.split('<script data-n-head="ssr" data-hid="page-schemas" type="application/ld+json">{')[1]

            # articleBody
            before, mk1, after = html.partition('"articleBody": "')
            articleBody, mk2, after = after.partition('",')
            articleBody = " ".join(articleBody[:].lstrip().split())

            # new_url
            before, mk1, after = html.partition('"text": "')
            new_url, mk2, after = after.partition('",')
            new_url = " ".join(new_url[:].lstrip().split())

            # dateline
            before, mk1, after = html.partition('"dateline": "')
            dateline, mk2, after = after.partition('",')
            dateline = " ".join(dateline[:].lstrip().split())

            # articleSection
            before, mk1, after = html.partition('"articleSection": "')
            articleSection, mk2, after = after.partition('",')
            articleSection = " ".join(articleSection[:].lstrip().split())

            # datePublished
            before, mk1, after = html.partition('"datePublished": "')
            datePublished, mk2, after = after.partition('",')
            datePublished = " ".join(datePublished[:].lstrip().split())

            # headline
            before, mk1, after = html.partition('"headline": "')
            headline, mk2, after = after.partition('",')
            headline = " ".join(headline[:].lstrip().split())

            # genre
            before, mk1, after = html.partition('"genre": "')
            genre, mk2, after = after.partition('",')
            genre = " ".join(genre[:].lstrip().split())

            # name
            before, mk1, after = html.partition('"name": "')
            name, mk2, after = after.partition('",')
            name = " ".join(name[:].lstrip().split())

            df.at[i,'url_2'] = new_url  
            df.at[i,'articleBody'] = " ".join(articleBody[:].lstrip().split())
            df.at[i,'dateline'] = dateline
            df.at[i,'headline'] = headline
            df.at[i,'articleSection'] = articleSection
            df.at[i,'datePublished'] = datePublished
            df.at[i,'genre'] = genre
            df.at[i,'name'] = name
  ######## 8 - try
          except:
            continue

        ############################################################################################################
        # this section is for the alternative broken URLS that use bit.ly but return 404 and need ammending
        ############################################################################################################

  ###### 6 - if response.status_code != 200:
        else: 
          try:
            try:
              ####################################################
              # This is for urls that have the title at as the 5th slash but the wrong front and end. eg:
              # http://www.forexlive.com/163470/all/ten-reasons-for-economic-optimism
              ####################################################
              wd.get(url)
              html = wd.page_source
              before, mk1, after = html[:1500].partition('og:url')
              subStr, mk2, after = after.partition(">")
              old_url = " ".join(subStr[:].lstrip().split())[11:-1]
              new_url_f = 'https://www.forexlive.com/news/!/'
              date = df.loc[i]['date']
              new_url = new_url_f + old_url.split('/')[5] +'-'+date.replace('-', '')
              df.at[i,'url_2'] = new_url
              wd.get(new_url)
              html = wd.page_source

              html = html.split('<script charset="utf-8" src="')[0]
              html = html.split('<script data-n-head="ssr" data-hid="page-schemas" type="application/ld+json">{')[1]

              # articleBody
              before, mk1, after = html.partition('"articleBody": "')
              articleBody, mk2, after = after.partition('",')
              articleBody = " ".join(articleBody[:].lstrip().split())

              # dateline
              before, mk1, after = html.partition('"dateline": "')
              dateline, mk2, after = after.partition('",')
              dateline = " ".join(dateline[:].lstrip().split())

              # articleSection
              before, mk1, after = html.partition('"articleSection": "')
              articleSection, mk2, after = after.partition('",')
              articleSection = " ".join(articleSection[:].lstrip().split())

              # datePublished
              before, mk1, after = html.partition('"datePublished": "')
              datePublished, mk2, after = after.partition('",')
              datePublished = " ".join(datePublished[:].lstrip().split())

              # headline
              before, mk1, after = html.partition('"headline": "')
              headline, mk2, after = after.partition('",')
              headline = " ".join(headline[:].lstrip().split())

              # genre
              before, mk1, after = html.partition('"genre": "')
              genre, mk2, after = after.partition('",')
              genre = " ".join(genre[:].lstrip().split())

              # name
              before, mk1, after = html.partition('"name": "')
              name, mk2, after = after.partition('",')
              name = " ".join(name[:].lstrip().split())

              df.at[i,'articleBody'] = " ".join(articleBody[:].lstrip().split())
              df.at[i,'dateline'] = dateline
              df.at[i,'headline'] = headline
              df.at[i,'articleSection'] = articleSection
              df.at[i,'datePublished'] = datePublished
              df.at[i,'genre'] = genre
              df.at[i,'name'] = name
            except:
              ####################################################
              # This is for urls that have the title at as the 7th slash but the wrong front and end. eg:
              # https://www.forexlive.com/blog/2014/01/02/good-start-for-portuguese-italian-and-spanish-debt-eurusd-technical-analysis-2-january-2014/
              ####################################################
              wd.get(url)
              html = wd.page_source
              before, mk1, after = html[:1500].partition('og:url')
              subStr, mk2, after = after.partition(">")
              old_url = " ".join(subStr[:].lstrip().split())[11:-1]
              new_url_f = 'https://www.forexlive.com/news/!/'
              date = df.loc[i]['date']
              new_url = new_url_f + old_url.split('/')[7] +'-'+date.replace('-', '')
              df.at[i,'url_2'] = new_url
              wd.get(new_url)
              html = wd.page_source

              html = html.split('<script charset="utf-8" src="')[0]
              html = html.split('<script data-n-head="ssr" data-hid="page-schemas" type="application/ld+json">{')[1]

              # articleBody
              before, mk1, after = html.partition('"articleBody": "')
              articleBody, mk2, after = after.partition('",')
              articleBody = " ".join(articleBody[:].lstrip().split())

              # dateline
              before, mk1, after = html.partition('"dateline": "')
              dateline, mk2, after = after.partition('",')
              dateline = " ".join(dateline[:].lstrip().split())

              # articleSection
              before, mk1, after = html.partition('"articleSection": "')
              articleSection, mk2, after = after.partition('",')
              articleSection = " ".join(articleSection[:].lstrip().split())

              # datePublished
              before, mk1, after = html.partition('"datePublished": "')
              datePublished, mk2, after = after.partition('",')
              datePublished = " ".join(datePublished[:].lstrip().split())

              # headline
              before, mk1, after = html.partition('"headline": "')
              headline, mk2, after = after.partition('",')
              headline = " ".join(headline[:].lstrip().split())

              # genre
              before, mk1, after = html.partition('"genre": "')
              genre, mk2, after = after.partition('",')
              genre = " ".join(genre[:].lstrip().split())

              # name
              before, mk1, after = html.partition('"name": "')
              name, mk2, after = after.partition('",')
              name = " ".join(name[:].lstrip().split())

              df.at[i,'articleBody'] = " ".join(articleBody[:].lstrip().split())
              df.at[i,'dateline'] = dateline
              df.at[i,'headline'] = headline
              df.at[i,'articleSection'] = articleSection
              df.at[i,'datePublished'] = datePublished
              df.at[i,'genre'] = genre
              df.at[i,'name'] = name
          except: #  fail to scrape anythion from url
############ 12
            try:
              ####################################################
              # This is for urls that have the title at as the 7th slash but the wrong front and end. eg:
              # https://www.forexlive.com/blog/2014/01/02/good-start-for-portuguese-italian-and-spanish-debt-eurusd-technical-analysis-2-january-2014/
              ####################################################
              wd.get(url)
              html = wd.page_source
              before, mk1, after = html[:1500].partition('og:url')
              subStr, mk2, after = after.partition(">")
              old_url = " ".join(subStr[:].lstrip().split())[11:-1]
              new_url_f = 'https://www.forexlive.com/news/!/'
              date = df.loc[i]['date']
              new_url = new_url_f + old_url.split('/')[4] +'-'+date.replace('-', '')
              df.at[i,'url_2'] = new_url
              wd.get(new_url)
              html = wd.page_source

              html = html.split('<script charset="utf-8" src="')[0]
              html = html.split('<script data-n-head="ssr" data-hid="page-schemas" type="application/ld+json">{')[1]

              # articleBody
              before, mk1, after = html.partition('"articleBody": "')
              articleBody, mk2, after = after.partition('",')
              articleBody = " ".join(articleBody[:].lstrip().split())

              # dateline
              before, mk1, after = html.partition('"dateline": "')
              dateline, mk2, after = after.partition('",')
              dateline = " ".join(dateline[:].lstrip().split())

              # articleSection
              before, mk1, after = html.partition('"articleSection": "')
              articleSection, mk2, after = after.partition('",')
              articleSection = " ".join(articleSection[:].lstrip().split())

              # datePublished
              before, mk1, after = html.partition('"datePublished": "')
              datePublished, mk2, after = after.partition('",')
              datePublished = " ".join(datePublished[:].lstrip().split())

              # headline
              before, mk1, after = html.partition('"headline": "')
              headline, mk2, after = after.partition('",')
              headline = " ".join(headline[:].lstrip().split())

              # genre
              before, mk1, after = html.partition('"genre": "')
              genre, mk2, after = after.partition('",')
              genre = " ".join(genre[:].lstrip().split())

              # name
              before, mk1, after = html.partition('"name": "')
              name, mk2, after = after.partition('",')
              name = " ".join(name[:].lstrip().split())

              df.at[i,'articleBody'] = " ".join(articleBody[:].lstrip().split())
              df.at[i,'dateline'] = dateline
              df.at[i,'headline'] = headline
              df.at[i,'articleSection'] = articleSection
              df.at[i,'datePublished'] = datePublished
              df.at[i,'genre'] = genre
              df.at[i,'name'] = name
############ 12
            except: # fail to scrape
              continue
###### 6   
      except: #  fail request response
        continue


## 2 - def

  # save the df as a csv
  df[['created_at',	'date',	'time',	'timezone',	'tweet',	'urls',	'articleBody',	'url_2',	'dateline',	'headline',	'articleSection',	'datePublished',	'genre',	'name']].to_csv('scraped_'+csv[6:-4]+'.csv', sep=',', index=False)

  # download saved csv to files
  files.download('scraped_'+csv[6:])

  # https://stackoverflow.com/questions/54293537/is-there-a-way-to-make-colab-give-an-audio-notification-when-cell-has-finished-r
  output.eval_js('new Audio("https://upload.wikimedia.org/wikipedia/commons/0/05/Beep-09.ogg").play()')
```

# Fix Grammar of articleBody using RegEx (yearly)
---



```python
'''
The concatination of bullet points into text sometimes causes gramatical errors int eh text. To ensure BERT does not get confued, we need to fix these using regular expression:
eg 
    "GBPJPY.Looking" to "GBPJPY. Looking"
    "fixingBoth" to "fixing. Both"
    "$-. 15" to "$-.15"
    "That is up $. 23 or 0.36%" to "That is up $.23 or 0.36%."

input:
  'forexlive_2022.csv'

output:
  'forexlive_2022.csv'

'''


import pandas as pd
from google.colab import files, output
import re

csv = ('forexlive_2021.csv')
df = pd.read_csv(csv)

def fix_parsing_errors(articleBody):
    """
      in:  Buyers in USDJPY and GBPUSD into the London fixingBoth the USDJPY and the GBPUSD traded to new highs into the London fixing, and that combination is a bullish cocktail for the GBPJPY.Looking at the hourly chart, the GBPJPY moved
      out: Buyers in USDJPY and GBPUSD into the London fixing. Both the USDJPY and the GBPUSD traded to new highs into the London fixing, and that combination is a bullish cocktail for the GBPJPY. Looking at the hourly chart, the GBPJPY moved
    """
    articleBody = re.sub(r'(?<=[a-z])(?=[A-Z0-9])','. ', articleBody) # eg. "fixingBoth" to "fixing. Both"
    articleBody = re.sub(r'(\d+\.\d+|\b[A-Z](?:\.[A-Z])*\b\.?)|([.,;:!?)])\s*', lambda x: x.group(1) or f'{x.group(2)} ', articleBody) # eg. "GBPJPY.Looking" to "GBPJPY. Looking"
    articleBody = re.sub(r'(?<=[$][.])([ ])(?=[A-Z0-9])',"", articleBody) # eg. "That is up $. 23 or 0.36%" to "That is up $.23 or 0.36%."
    articleBody = re.sub(r'(?<=[$][-][.])([ ])(?=[A-Z0-9])',"", articleBody) # eg. "$-. 15" to "$-.15"
    articleBody = articleBody.replace('%', "% ")
    articleBody = articleBody.replace('%', "% ")
    articleBody = articleBody.replace('0k', "0k. ")
    articleBody = articleBody.replace('0K', "0K. ")
    articleBody = articleBody.replace('0m', "0m. ")
    articleBody = articleBody.replace('0M', "0M. ")
    articleBody = articleBody.replace('0b', "0b. ")
    articleBody = articleBody.replace('0B', "0B. ")
    articleBody = articleBody.replace('0t', "0t. ")
    articleBody = articleBody.replace('0T', "0T. ")
    articleBody = articleBody.replace('  ', " ")
    articleBody = articleBody.replace(' , ', ", ")
    articleBody = articleBody.replace(' . ', ". ")
    return str(articleBody)

for i in range(1,len(df)):
  try:
    articleBody = df['articleBody'][i]
    # print(i, articleBody)
    df.at[i,'articleBody'] = fix_parsing_errors(articleBody)
  except:
    continue

# # after check print
# i = 23
# print(df['articleBody'][i])

# # # print(fix_parsing_errors(articleBody))
df.to_csv(csv, sep=',', index=False)
files.download(csv)
```

# FinBERT sentiment analysis of headline and articleBody (yearly)

```python
'''
load the FinBERT tokenizer and model as well as the required packages
'''

!pip install transformers
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from tqdm import tqdm
from google.colab import files, output

#Get the tokenizer
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")

#Get the model
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

```python
'''
this takes the cleaned csv with article body and headline data, then applies sentiment analysis to them, adding a score to a pos_sent, neg_sent and neu_sent columns

input:
  'forexlive_2022.csv'
  etc

output:
  'sent_forexlive_2022.csv'
  etc


the output of this cell is ready to be used int he model and is places d int eh 'forexlive', '4_finbert' folder, which is accessed by feature_engineer.py
'''

csv = ('forexlive_2021.csv')

df = pd.read_csv(csv)
df['pos_sent'] = float()
df['neg_sent'] = float()
df['neu_sent'] = float()

for i in tqdm(range(len(df))):
  try:
    combined_text = []
    combined_text.append(str(df['headline'][i]+'. '+df['articleBody'][i]))
    # print(combined_text)
    tokenized_combined_text = tokenizer(combined_text, padding = True, truncation = True, return_tensors='pt')
    outputs = model(**tokenized_combined_text)
    predict = torch.nn.functional.softmax(outputs.logits, dim=-1)
    model.config.id2label
    df.at[i,'pos_sent'] = predict[:, 0]
    df.at[i,'neg_sent'] = predict[:, 1]
    df.at[i,'neu_sent'] = predict[:, 2]
  except:
    continue

# send to csv and download
df.to_csv('sent_'+csv, sep=',', index=False)
files.download('sent_'+csv)

# https://stackoverflow.com/questions/54293537/is-there-a-way-to-make-colab-give-an-audio-notification-when-cell-has-finished-r
output.eval_js('new Audio("https://upload.wikimedia.org/wikipedia/commons/0/05/Beep-09.ogg").play()')
```
